```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#example of chain codes
# Load the image and convert to grayscale

img = cv2.imread('image1.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Draw the contour on original image
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0,255,0), 2)

plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title('Detected Contour')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image1.png', cv2.IMREAD_GRAYSCALE)
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Grayscale')
plt.subplot(1,2,2); plt.imshow(binary, cmap='gray'); plt.title('Binary')
plt.show()
```

```python
# Define the structuring element (kernel)
kernel = np.ones((5, 5), np.uint8)

# Apply erosion
erosion = cv2.erode(binary, kernel, iterations=1)

# Display erosion result
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')
plt.axis('off')
plt.show()
```

```python
# Apply dilation
dilation = cv2.dilate(binary, kernel, iterations=1)

# Display dilation result
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')
plt.axis('off')
plt.show()
```

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

# Display opening result
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')
plt.show()
```

```python
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)

# Display closing result
plt.imshow(closing, cmap='gray')
plt.title('Closing')
plt.axis('off')
plt.show()
```

```python
num_labels, labels = cv2.connectedComponents(binary)
# Visualize labels
plt.imshow(labels, cmap='nipy_spectral')
plt.title(f"Connected Components: {num_labels -1 } objects")
plt.axis('off')
plt.show()
```
